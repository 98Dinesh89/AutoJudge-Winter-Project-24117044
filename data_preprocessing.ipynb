```python
import pandas as pd
import numpy as np
from scipy.sparse import hstack, save_npz

df = pd.read_json("data/problems_data.jsonl", lines=True)
```

```python
df["full_text"] = (
    df["title"].fillna("") + " " +
    df["description"].fillna("") + " " +
    df["input_description"].fillna("") + " " +
    df["output_description"].fillna("")
)

df["text_length"] = df["full_text"].apply(len)
```

```python
math_symbols = ['$', '<', '>', '=', '+', '-', '*', '/', '^']

def count_math_symbols(text):
    return sum(text.count(sym) for sym in math_symbols)

df["math_symbol_count"] = df["full_text"].apply(count_math_symbols)
```

```python
df_final = df[
    [
        "full_text",
        "text_length",
        "math_symbol_count",
        "problem_class",
        "problem_score" 
    ]
]
```
